Change the inputs in the following cell:

```python
stations = ['FISH', 'MIDG', 'TRIN', 'BOLI', 'OLDR', '8771486','8770613','8771013','8771341','8770777']

startdate = '2019-9-1'
enddate = '2019-10-1'
```

# Read in Galveston Bay data from TABS website

You can change any of the code in these cells, and add your own cells and do more things. 

Because data from different stations is gathered at different rates, it is all interpolated to 15 minutes.

To run this notebook, click

> Cell > Run All

After it has finished running, the data can be found on the tab to the left called "Home". Click on that tab, click on the little box in front of `data.csv` and then select "Download" to download the data to your computer.

Change the inputs in the following cell:

```python
# List of stations in Galveston Bay
stations = ['FISH', 'MIDG', 'TRIN', 'BOLI', 'OLDR', '8771486','8770613','8771013','8771341','8770777','g06010']

# 'year-month-day' for start and end dates
startdate = '2017-9-1'
enddate = '2017-10-1'
```

```python
# you can ignore the warnings!

df = pd.DataFrame()

# loop over stations to read in data
for station in stations:
    dft = tabs.read(station, startdate, enddate, resample=('15T', 0, 'instant', False))
    df = pd.concat([df,dft],axis=1)
```
